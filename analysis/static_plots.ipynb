Based on the plots above it makes more sense to look at popularity, energy vs. loudness, danceability vs. explicitly (a categorical variable), and valence with mode (a categorical variable)

Also, for the correlation maps simplicity will use only danceability, energy, valence, popularity, and tempo. A correlation map with all the continuous variables will be generated at the end. 

```python
figure_1 = plt.figure(constrained_layout = True, figsize = (14, 8)) # Defining the figure and layout
grid_spec = figure_1.add_gridspec(2, 3)

ax1 = figure_1.add_subplot(grid_spec[:, 0]) # The first ax object will be a heatmap
correlation_map = dataframe_filtered[['danceability', 'energy', 'valence', 'popularity', 'tempo']].corr() # The heat map with reduced variables to fit in
sns.heatmap(correlation_map, annot=True, cmap='coolwarm', ax=ax1) # This will have annotations and a good color palete that is perceptually uniform
ax1.set_title('Correlation Heatmap') # The title of the individual ax object

# Next, will be a histogram showing popularity
ax2 = figure_1.add_subplot(grid_spec[0, 1])
dataframe_filtered_new = dataframe_filtered[dataframe_filtered['popularity'] != 0] # Removing popularity scores equal to zero
sns.histplot(dataframe_filtered_new['popularity'], kde=True, ax=ax2, color='plum') # Used is a color that is not distractive
ax2.set_title('Popularity Distribution Histogram') # Set the title of the plot
popularity_mean = dataframe_filtered_new['popularity'].mean() # Calculates the mean score for popularity
ax2.axvline(popularity_mean, color='black', linestyle='--') # Plotting this value on the plot with a dashed line and line style of --
ax2.text(popularity_mean - 25, ax2.get_ylim()[1]*0.8, f'Mean: {popularity_mean:.1f}', color='black') # Display the mean value on the chart without disturbing the plot, on plot annotation

# Next, a box plot of danceability and explicity. This is a categorical variable vs. a quantitative one. 
ax3 = figure_1.add_subplot(grid_spec[1, 1]) # Will appear below the histogram
# Will use the original dataframe with popularity values of zero. Will remove the legend since it is quite obvious
sns.boxplot(data = dataframe_filtered, x = 'explicit', y = 'danceability', ax = ax3, hue = 'explicit', palette = 'gray', legend = False) # Used a gray scale to avoid confusion among color blind
ax3.set_title('Box plot of Danceability vs Explicit') # The title

ax4 = figure_1.add_subplot(grid_spec[0, 2])
# Reduced the size of dots to make it appear more clearer
sns.scatterplot(data = dataframe_filtered, x = 'loudness', y = 'energy', hue = 'explicit', ax = ax4, alpha = 0.6, s = 5) # These will be color based on explicity
ax4.set_title('Scatter plot of Energy vs Loudness') # Next a scatter plot focusing on energy and loudness
ax4.legend(title='Explicit', loc='lower right') # Gives a legend to indicate color depends on explicitly to avoid confusion
ax4.tick_params(axis = 'x', labelsize = 8) # Reduced the size of tick labels a bit to make it less crammed up 

# Next, a voilin plot was used on mode and valence
ax5 = figure_1.add_subplot(grid_spec[1, 2]) # This will appear below the scatter plot
# This will be colored based on the mode so the legend was removed since this was unnecessary
sns.violinplot(data = dataframe_filtered, x = 'mode', y = 'valence', ax = ax5, hue = 'mode', palette = 'coolwarm', legend = False) # A perceptually uniform palette was used
ax5.set_title('Voilin Plot of Valence by Mode')

plt.suptitle('Visual Exploration', fontsize=16) # Super title
plt.show() # Starts the plot
```

- The popularity is still left skewed since the mean is less than the median.

- Explicit songs scored better for danceability compared to nonexplicit songs.

- Higher energy songs tend to be louder however this plot does not seem to have a clear correlation of how explicitly affects this.

- Songs with a mode = 1, have lower scores for valence, while songs with a mode = 0, have higher scores for valence. This might not make much sense in the actual world. However, this means that major key songs trend toward higher valence, while minor ones have a lower valence. 

```python
figure_2 = plt.figure(constrained_layout = True, figsize = (8, 6)) # Using the object-oriented approach to do a on plot labeling
grid_spec = figure_2.add_gridspec(1, 1)
ax1 = figure_2.add_subplot(grid_spec[:,:]) # Plots the entire gridspec
# Increased the size of dots to make it appear clearer, and coloring will be based on the year, the Viridis palette is used the opacity was decreased
sns.scatterplot(data = dataframe_filtered, x = 'loudness', y = 'energy', hue = 'year', palette='viridis', ax = ax1, alpha = 0.4, s = 20) 
ax1.legend(title = 'Year', loc = 'lower right') # Gives a legend to indicate color depends on year to avoid confusion
ax1.tick_params(axis = 'x', labelsize = 8) # Reduced the size of tick labels a bit to make it less crammed up 

# Annotation, The arrow points to xy, and the arrow starts from xytext, showing where there are more data points
ax1.annotate('More data points', xy = (-4.5, 0.8), xytext = (-16, 0.95), arrowprops = dict(facecolor = 'black', shrink = 2), fontsize = 10)
# Annotation points to where there are fewer data points
ax1.annotate('Less data points', xy = (-17, 0.2), xytext = (-19, 0.5),  size = 10, arrowprops = {'facecolor': '#fcba03', 'shrink': 2}, bbox = {'boxstyle': 'round', 'color': 'gray'})

plt.suptitle('Scatter plot of Energy vs Loudness color based on year', fontsize = 16) # Super title
plt.show() # Starts the plot
```

# Do distributions or central tendencies of continuous variables differ across levels of categorical variables? 

- Valence by mode already seems to have a strong relationship

Next danceability was compared with the most common genres. The top 10 genres were considered.

```python
plt.figure(figsize = (12, 6)) # Defines the plot size
top_genres = dataframe_filtered['genre']
top_genres = top_genres.value_counts()
top_genres = top_genres.nlargest(10) # Picks the 10 with the most common values
top_genres = top_genres.index # The index of the top genres

sns.boxplot(x = 'genre', y = 'danceability', color = 'gray', data = dataframe_filtered[dataframe_filtered['genre'].isin(top_genres)])
plt.title('Box plot of Danceability Plotted with each Genre')
plt.xticks(rotation = 45) # Rotated ticks to make them more visible
plt.ylabel('Danceability')
plt.xlabel('Genre')
plt.tight_layout() # Used the tight layout to optimize the space
plt.show()
```

Some genres appear to be more danceable than others. It can be seen songs classified "hip hop, pop, R & B", and just "hip hop" and "hip hop, pop" seem to be more danceable when making comparisons from the 25th to the 75th percentile which is the IQR.
